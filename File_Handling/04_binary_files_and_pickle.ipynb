```python
# Practice 1 – Write Simple Binary Data:
with open("sample.bin", "wb") as file:
    file.write(b"Hello Binary")   # write bytes data

print("Binary file created")
```

```python
# Practice 2 – Read Binary Data:
with open("sample.bin", "rb") as file:
    data = file.read()   # read bytes

print(data)              # raw bytes
print(type(data))        # should be bytes
```

```python
# Practice 3 – Compare Text vs Binary:
# Text mode

with open("compare.txt", "w") as file:
    file.write("Hello Giri")

with open("compare.txt", "r") as file:
    text_data = file.read()

print(text_data)
print(type(text_data))   # str
```

```python
# Binary mode

with open("compare.bin", "wb") as file:
    file.write(b"Hello Giri")   # bytes

with open("compare.bin", "rb") as file:
    binary_data = file.read()

print(binary_data)
print(type(binary_data))   # bytes
```

```python
# Practice 4 – Save Dictionary Using Pickle:
import pickle

student = {"name": "Giri", "age": 27}

with open("student.pkl", "wb") as file:
    pickle.dump(student, file)   # save object

print("Dictionary saved")
```

```python
# Practice 5 – Load Dictionary Back:
import pickle

with open("student.pkl", "rb") as file:
    loaded_student = pickle.load(file)   # load object

print(loaded_student)
print(type(loaded_student))
```

```python
# Practice 6 – Save List Using Pickle:
import pickle

numbers = [10, 20, 30, 40]

with open("numbers.pkl", "wb") as file:
    pickle.dump(numbers, file)   # save list

print("List saved")
```

```python
# Practice 7 – Load List Back:
import pickle

with open("numbers.pkl", "rb") as file:
    loaded_numbers = pickle.load(file)

print(loaded_numbers)
```

```python
# Practice 8 – Save Multiple Objects:
import pickle

data = {"course": "AI/ML", "completed": False}

with open("multi.pkl", "wb") as file:
    pickle.dump(data, file)
    pickle.dump([1, 2, 3], file)   # save second object

print("Multiple objects saved")
```

```python
# Practice 9 – Load Multiple Objects:
import pickle

with open("multi.pkl", "rb") as file:
    obj1 = pickle.load(file)
    obj2 = pickle.load(file)

print(obj1)
print(obj2)
```
