# File Reading Methods:

```python
# Step 1 – Setup File : Create a file with multiple lines.
with open("data.txt", "w") as file:
    file.write("Line 1\n")
    file.write("Line 2\n")
    file.write("Line 3\n")
    file.write("Line 4\n")

print("File prepared.")
```

```python
# Step 2 – read():
with open("data.txt", "r") as file:
    content = file.read()
    print(content)

# Reads entire file at once

# Returns single string
```

```python
# Step 3 – readline():

with open("data.txt", "r") as file:
    print(file.readline())
    print(file.readline())

# Reads one line at a time

# Pointer moves forward after each call
```

```python
# Step 4 – readlines():
with open("data.txt", "r") as file:
    lines = file.readlines()

print(lines)

# Returns list

# Each line is an element

# \n included
```

```python
# Step 5 – Loop Through File:
with open("data.txt", "r") as file:
    for line in file:
        print(line.strip())
```

### Practice Exercises:

```python
# Practice 1 – Read Entire File:
with open("data1.txt", "r") as file:
    print(file.read())
```

```python
# Practice 2 – Read Line by Line:
with open("data1.txt", "r") as file:
    print(file.readline())
    print(file.readline())
```

```python
# Practice 3 – Read All Lines as List:
with open("data1.txt", "r") as file:
    lines = file.readlines()

print(lines)
```

```python
# Practice 4 – Loop Through File:
with open("data1.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
# Practice 5 – Pointer Moves Line by Line:

with open("data1.txt", "r") as file:
    print(file.readline())   # reads first line
    print(file.readline())   # reads second line (pointer moved)
```

```python
# Practice 6 – Second read() Returns Empty:

with open("data1.txt", "r") as file:
    print(file.read())       # reads full file
    
    print("Second read:")
    print(file.read())       # empty because pointer is at end
```

```python
# Practice 7 – Reset Pointer Using seek():

with open("data1.txt", "r") as file:
    print(file.read())       # pointer goes to end
    
    file.seek(0)             # reset pointer to beginning
    
    print("After reset:")
    print(file.read())       # reads again from start
```
