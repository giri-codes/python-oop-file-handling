````python
# Practice 1 – Check If File Exists:
import os
if os.path.exists("data.txt"):   # check file existence
    print("File exists")

else:
    print("File does not exist")
````

````python
# Practice 2 – Delete a File:
import os

if os.path.exists("temp.txt"):
    os.remove("temp.txt")   # delete file
    print("File deleted")
else:
    print("File not found")
````

````python
# Practice 3 – Rename a File:
import os

if os.path.exists("data.txt"):
    os.rename("data.txt", "new_data.txt")   # rename file
    print("File renamed")
else:
    print("File not found")
````

````python
# Practice 4 – Create a Folder:
import os

if not os.path.exists("my_folder"):
    os.mkdir("my_folder")   # create folder
    print("Folder created")
else:
    print("Folder already exists")
````

````python
# Practice 5 – Delete a Folder:
import os

if os.path.exists("my_folder"):
    os.rmdir("my_folder")   # remove empty folder
    print("Folder removed")
else:
    print("Folder not found")
````

````python
# Practice 6 – List Files in Current Directory:
import os

for item in os.listdir():
    print(item)
````

````python
# Practice 7 – Safe Rename (Prevent Error):
import os

if os.path.exists("old.txt") and not os.path.exists("new.txt"):
    os.rename("old.txt", "new.txt")   # rename safely
    print("Renamed successfully")
else:
    print("Rename not possible")
````

````python
# Practice 8 – Create Folder and File Inside It:
import os

if not os.path.exists("project"):
    os.mkdir("project")   # create folder

file_path = os.path.join("project", "info.txt")

with open(file_path, "w") as file:
    file.write("Project file created")

print("Folder and file ready")
````
