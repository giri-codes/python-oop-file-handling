# Inner Class:

### Step 1 – What is an Inner Class?

An inner class is a class defined inside another class.

```python
# Example 1: Laptop and Processor (Inner Class)

class Laptop:
    def __init__(self, brand):
        self.brand = brand
        self.processor = self.Processor()

    def show(self):
        print("Laptop brand:", self.brand)

    class Processor:
        def details(self):
            print("Processor: Intel i7")

l = Laptop("HP")
l.show()
l.processor.details()
```

```python
# Example 2: University and Department

class University:
    def __init__(self, name):
        self.name = name
        self.department = self.Department()

    def show(self):
        print("University:", self.name)

    class Department:
        def info(self):
            print("Department: Computer Science")

u = University("VTU")
u.show()
u.department.info()
```

```python
# Example 3: Bank and Account

class Bank:
    def __init__(self, bank_name):
        self.bank_name = bank_name
        self.account = self.Account()

    def show_bank(self):
        print("Bank:", self.bank_name)

    class Account:
        def account_type(self):
            print("Account Type: Savings")

b = Bank("SBI")
b.show_bank()
b.account.account_type()
```

```python
# Example 4: Company and Employee

class Company:
    def __init__(self, name):
        self.name = name
        self.employee = self.Employee()

    def show_company(self):
        print("Company:", self.name)

    class Employee:
        def role(self):
            print("Role: Software Engineer")

c = Company("Infosys")
c.show_company()
c.employee.role()
```

## Step 2 – Inner Class Accessing Outer Class Data

```python
# Example 1: School and Classroom

class School:
    def __init__(self, name):
        self.name = name
        self.classroom = self.Classroom(self)

    class Classroom:
        def __init__(self, outer):
            self.outer = outer

        def show_school(self):
            print("School Name:", self.outer.name)

s = School("Oxford School")
s.classroom.show_school()
```

```python
# Example 2: Hospital and Doctor

class Hospital:
    def __init__(self, hospital_name):
        self.hospital_name = hospital_name
        self.doctor = self.Doctor(self)

    class Doctor:
        def __init__(self, outer):
            self.outer = outer

        def details(self):
            print("Doctor works at", self.outer.hospital_name)

h = Hospital("Apollo")
h.doctor.details()
```

```python
# Example 3: Company and Department

class Company:
    def __init__(self, company_name):
        self.company_name = company_name
        self.department = self.Department(self)

    class Department:
        def __init__(self, outer):
            self.outer = outer

        def info(self):
            print("Department belongs to", self.outer.company_name)

c = Company("Infosys")
c.department.info()
```

```python
# Example 4: College and Student

class College:
    def __init__(self, college_name):
        self.college_name = college_name
        self.student = self.Student(self)

    class Student:
        def __init__(self, outer):
            self.outer = outer

        def show_college(self):
            print("Student studies at", self.outer.college_name)

col = College("BRT College")
col.student.show_college()
```
