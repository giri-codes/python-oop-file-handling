# PYTHON OOPS:
Oop(`Object Oriented Programming`) is a way of writing code by thinking in terms of real world objects.

```python
# PRACTICE – CLASSES & OBJECTS:
# Ex-1: Empty class:
class Car:
    pass

c1 = Car()
c2 = Car()
```

```python
# Ex-2 Class with variables:
class Student:
    name = "Giri"
    age = 25

s1 = Student()
print(s1.name)
print(s1.age)
```

```python
# Ex-3 Multiple objects:
s2 = Student()
print(s2.name)
```

```python
# Ex-4
class Bike:
    brand = "Jawa"

b1 = Bike()
print(b1.brand)
```

```python
# PRACTICE – METHODS:
# Ex-1 
class Student:
    def study(self):
        print("Studying python")

s1 = Student()
s1.study()
```

```python
# Ex-2
class Dog:
    def bark(self):
        print("Dog is baking")

d1 = Dog()
d1.bark()
```

```python
# Ex-3 Method using class variable:
class College:
    name = "C E S"
    def show_name(self):
        print(self.name)

c1 = College()
c1.show_name()
```

```python
# __init__() CONSTRUCTOR:
# Ex-1: Basic constructor
class District:
    def __init__(self, name):
        self.name = name

d1 = District("Mysuru")
print(d1.name)
```

```python
# Ex-2 Two parameters:
class Employee:
    def __init__(self, name, age):
        self.name = name
        self.age = age

e1 = Employee("Giri", 21)
e2 = Employee("Akash", 22)

print(e1.name, e1.age)
print(e2.name, e2.age)
```

```python
# Ex - 3
class Mobile:
    def __init__(self, brand):
        self.brand = brand

m1 = Mobile("Samsung")
m2 = Mobile("Apple")

print(m1.brand)
print(m2.brand)
```

```python
# Practice 1: class & object:
class Book:
    title = "Python Basics"
    pages = 300

b1 = Book()
b2 = Book()

print(b1.title)
print(b2.title)
```

```python
# practice 2: Methods
class Fan:
    def turn_on(self):
        print("Fan is ON")

    def turn_off(self):
        print("Fan is OFF")

f1 = Fan()
f1.turn_on()
f1.turn_off()
```

```python
# Practice 3 : __init__ Constructor:
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

e1 = Employee("Giri", 50000)
e2 = Employee("Ravi", 40000)

print(e1.name, e1.salary)
print(e2.name, e2.salary)
```

```python
# Practice 4 :

class Laptop:
    def __init__(self, brand):
        self.brand = brand

    def show(self):
        print(self.brand)

l1 = Laptop("HP")
l1.show()
```

```python
# Ex-1
class ClassName:
    def __init__(self, data1, data2):
        self.data1 = data1
        self.data2 = data2

    def action1(self):
        print(self.data1)
        
    def action2(self):
        print(self.data2)
c1 = ClassName("9th std", "10th std")
c1.action1()
c1.action2()
```

```python
# Ex-2
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def study(self):
        print(self.name, "is studying")

    def show_age(self):
        print("Age:", self.age)

s2 = Student("giri", 24)
s2.study()
s2.show_age()
```
